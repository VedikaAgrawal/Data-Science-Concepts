# Clustering

### 1. What is unsupervised learning in the context of machine learning?
[cite_start]Unsupervised learning is a branch of machine learning where the model is trained on data that does not have pre-defined labels or target outcomes[cite: 9]. [cite_start]The goal is to discover underlying patterns, structures, or groupings within the data without explicit guidance[cite: 9].

### 2. How does K-Means clustering algorithm work?
[cite_start]K-Means is a centroid-based algorithm that partitions data into $K$ clusters[cite: 10]. It follows these steps:
* [cite_start]**Initialization:** $K$ points are randomly selected as initial centroids[cite: 10].
* [cite_start]**Assignment:** Each data point is assigned to the nearest centroid based on a distance metric (usually Euclidean distance)[cite: 10].
* [cite_start]**Update:** The centroid of each cluster is recalculated by taking the mean of all points assigned to it[cite: 10].
* [cite_start]**Convergence:** These steps repeat until centroids stabilize or a maximum number of iterations is reached[cite: 10].

### 3. Explain the concept of a dendrogram in hierarchical clustering.
[cite_start]A dendrogram is a tree-like diagram that records the sequences of merges or splits during hierarchical clustering[cite: 11]. [cite_start]The vertical axis represents the distance or dissimilarity between clusters, and the horizontal axis represents individual data points[cite: 11]. [cite_start]It allows users to visualize the relationships between clusters and decide where to "cut" the tree to obtain a specific number of clusters[cite: 11].

### 4. What is the main difference between K-Means and Hierarchical Clustering?
The primary differences are:
* [cite_start]**Requirement of K:** K-Means requires the number of clusters ($K$) to be specified upfront, whereas Hierarchical Clustering does not[cite: 12].
* [cite_start]**Structure:** K-Means produces a flat partition of data, while Hierarchical Clustering creates a nested, tree-like structure (dendrogram)[cite: 12].
* [cite_start]**Flexibility:** Hierarchical clustering allows you to choose the number of clusters after the algorithm has run by analyzing the dendrogram[cite: 12].

### 5. What are the advantages of DBSCAN over K-Means?
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) offers several benefits:
* [cite_start]**Arbitrary Shapes:** It can identify clusters of non-spherical and irregular shapes, unlike K-Means which assumes clusters are spherical[cite: 13].
* [cite_start]**Noise Handling:** It explicitly identifies and labels outliers as "noise" rather than forcing them into a cluster[cite: 13].
* [cite_start]**No Fixed K:** It does not require a predefined number of clusters[cite: 13].

### 6. When would you use Silhouette Score in clustering?
[cite_start]The Silhouette Score is used to evaluate the quality of clustering by measuring how similar an object is to its own cluster compared to other clusters[cite: 14]. You would use it to:
* [cite_start]Determine the optimal number of clusters[cite: 14].
* [cite_start]Assess the separation between clusters and the cohesion within them[cite: 14].

### 7. What are the limitations of Hierarchical Clustering?
* [cite_start]**Scalability:** It is computationally expensive ($O(n^2)$ or $O(n^3)$), making it unsuitable for very large datasets[cite: 15].
* [cite_start]**Irreversibility:** Once two clusters are merged, they cannot be separated in later steps[cite: 15].
* [cite_start]**Sensitivity:** It can be sensitive to noise and outliers depending on the linkage criteria used[cite: 15].

### 8. Why is feature scaling important in clustering algorithms like K-Means?
[cite_start]K-Means relies on distance calculations (Euclidean distance) to group points[cite: 16]. [cite_start]If features have different scales (e.g., age vs. annual income), features with larger numerical ranges will dominate the distance calculation, leading to biased results[cite: 16]. [cite_start]Scaling ensures all features contribute equally[cite: 16].

### 9. How does DBSCAN identify noise points?
[cite_start]DBSCAN identifies noise points by checking density requirements[cite: 17]. [cite_start]If a point is not a "core point" (having at least `min_samples` within `eps` distance) and is not within the neighborhood of any core point, it is labeled as noise[cite: 17].

### 10. Define inertia in the context of K-Means.
[cite_start]Inertia (or Within-Cluster Sum of Squares) is the sum of squared distances between each data point and its assigned cluster centroid[cite: 18]. [cite_start]It measures how tightly packed the clusters are; lower inertia generally suggests more compact clusters[cite: 18].

### 11. What is the elbow method in K-Means clustering?
[cite_start]The elbow method is a technique used to find the optimal value for $K$[cite: 19]. [cite_start]By plotting the inertia against the number of clusters, the "elbow" point—where the rate of decrease in inertia slows significantly—indicates the most appropriate number of clusters[cite: 19].

### 12. Describe the concept of "density" in DBSCAN.
[cite_start]In DBSCAN, density refers to the number of data points located within a specific radius (defined by the parameter `eps`) around a point[cite: 20]. [cite_start]A region is considered dense if it contains at least a minimum number of points (`min_samples`)[cite: 20].

### 13. Can hierarchical clustering be used on categorical data?
[cite_start]Yes, hierarchical clustering can be used on categorical data if an appropriate dissimilarity metric (such as Gower distance or Jaccard similarity) is used instead of standard Euclidean distance[cite: 21].

### 14. What does a negative Silhouette Score indicate?
[cite_start]A negative Silhouette Score indicates that a data point has likely been assigned to the wrong cluster, as it is closer to a neighboring cluster than to its own[cite: 22].

### 15. Explain the term "linkage criteria" in hierarchical clustering.
[cite_start]Linkage criteria define how the distance between two sets (clusters) of observations is calculated[cite: 23]. Common methods include:
* [cite_start]**Single Linkage:** Minimum distance between points in two clusters[cite: 23].
* [cite_start]**Complete Linkage:** Maximum distance between points in two clusters[cite: 23].
* [cite_start]**Average Linkage:** Average distance between all pairs of points in two clusters[cite: 23].

### 16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
[cite_start]K-Means aims to minimize the variance within clusters and assumes they are roughly spherical and similar in size[cite: 24]. [cite_start]If clusters have different densities or sizes, K-Means may incorrectly split larger clusters or merge smaller ones into larger ones to minimize global inertia[cite: 24].

### 17. What are the core parameters in DBSCAN, and how do they influence clustering?
* [cite_start]**Eps (ε):** The maximum distance between two samples for them to be considered as neighbors[cite: 25]. [cite_start]Smaller `eps` leads to more noise; larger `eps` merges clusters[cite: 25].
* [cite_start]**Min_samples:** The number of samples in a neighborhood for a point to be considered a core point[cite: 25]. [cite_start]Higher values require higher density to form a cluster[cite: 25].

### 18. How does K-Means++ improve upon standard K-Means initialization?
[cite_start]K-Means++ improves initialization by choosing initial centroids that are far apart from each other[cite: 26]. [cite_start]It selects the first centroid randomly and then selects subsequent centroids with a probability proportional to their squared distance from the nearest existing centroid, leading to faster convergence and better final results[cite: 26].

### 19. What is agglomerative clustering?
[cite_start]Agglomerative clustering is a "bottom-up" hierarchical approach[cite: 27]. [cite_start]Every data point starts as its own individual cluster, and in each step, the two most similar clusters are merged until only one cluster (or the desired number) remains[cite: 27].

### 20. What makes Silhouette Score a better metric than just inertia for model evaluation?
[cite_start]Inertia only measures how tight the clusters are and will always decrease as $K$ increases[cite: 28]. [cite_start]The Silhouette Score is superior because it considers both **cohesion** (how close points are within a cluster) and **separation** (how far clusters are from each other), providing a peak value that clearly indicates the best model fit[cite: 28].

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import (make_blobs, make_moons, make_circles, 
                              load_iris, load_wine, load_breast_cancer, load_digits)
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score, silhouette_samples
from scipy.cluster.hierarchy import dendrogram, linkage

# 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means.
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
kmeans = KMeans(n_clusters=4, n_init='auto')
y_kmeans = kmeans.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.title("21. K-Means with 4 Blobs")
plt.show()

# 22. Load Iris and use Agglomerative Clustering (3 clusters). Display first 10 predicted labels.
iris = load_iris()
agg = AgglomerativeClustering(n_clusters=3)
labels = agg.fit_predict(iris.data)
print(f"22. First 10 labels: {labels[:10]}")

# 23. Generate make_moons and apply DBSCAN. Highlight outliers.
X, _ = make_moons(n_samples=200, noise=0.1, random_state=0)
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='coolwarm')
plt.title("23. DBSCAN on Moons (Label -1 = Outlier)")
plt.show()

# 24. Load Wine dataset, apply K-Means after standardizing. Print size of each cluster.
wine = load_wine()
X_scaled = StandardScaler().fit_transform(wine.data)
kmeans = KMeans(n_clusters=3, n_init='auto', random_state=42).fit(X_scaled)
print("24. Cluster Sizes:\n", pd.Series(kmeans.labels_).value_counts())

# 25. Use make_circles to generate synthetic data and cluster it using DBSCAN.
X, _ = make_circles(n_samples=500, factor=0.5, noise=0.05)
dbscan = DBSCAN(eps=0.15, min_samples=5).fit(X)
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)
plt.title("25. DBSCAN on Circles")
plt.show()

# 26. Load Breast Cancer, apply MinMaxScaler, K-Means (2 clusters). Output centroids.
cancer = load_breast_cancer()
X_scaled = MinMaxScaler().fit_transform(cancer.data)
kmeans = KMeans(n_clusters=2, n_init='auto').fit(X_scaled)
print("26. Cluster Centroids:\n", kmeans.cluster_centers_)

# 27. make_blobs with varying cluster standard deviations and cluster with DBSCAN.
X, _ = make_blobs(n_samples=500, cluster_std=[0.5, 1.5, 0.7], random_state=42)
dbscan = DBSCAN(eps=0.8, min_samples=5).fit(X)
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)
plt.title("27. DBSCAN with Varying Densities")
plt.show()

# 28. Load Digits dataset, reduce to 2D using PCA, visualize K-Means.
digits = load_digits()
X_pca = PCA(n_components=2).fit_transform(digits.data)
kmeans = KMeans(n_clusters=10, n_init='auto').fit(X_pca)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='tab10', alpha=0.7)
plt.title("28. Digits K-Means (PCA 2D)")
plt.show()

# 29. make_blobs, evaluate silhouette scores for k=2 to 5. Display as bar chart.
X, _ = make_blobs(n_samples=500, centers=4, random_state=42)
k_range = range(2, 6)
scores = [silhouette_score(X, KMeans(n_clusters=k, n_init='auto').fit_predict(X)) for k in k_range]
plt.bar(k_range, scores)
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.title("29. Silhouette Scores")
plt.show()

# 30. Load Iris dataset and plot a dendrogram with average linkage.
linked = linkage(iris.data, 'average')
plt.figure(figsize=(10, 5))
dendrogram(linked)
plt.title("30. Iris Dendrogram (Average Linkage)")
plt.show()

# 31. Overlapping clusters (make_blobs), K-Means with decision boundaries.
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)
kmeans = KMeans(n_clusters=3, n_init='auto').fit(X)
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, edgecolors='k')
plt.title("31. Decision Boundaries")
plt.show()

# 32. Load Digits, apply DBSCAN after t-SNE.
X_tsne = TSNE(n_components=2, random_state=42).fit_transform(digits.data)
dbscan = DBSCAN(eps=4.0, min_samples=10).fit(X_tsne)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=dbscan.labels_, cmap='tab10')
plt.title("32. Digits DBSCAN (t-SNE)")
plt.show()

# 33. make_blobs, Agglomerative Clustering with complete linkage.
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)
agg = AgglomerativeClustering(n_clusters=3, linkage='complete').fit(X)
plt.scatter(X[:, 0], X[:, 1], c=agg.labels_)
plt.title("33. Complete Linkage")
plt.show()

# 34. Breast Cancer dataset, compare inertia for K=2 to 6. Line plot.
inertia = [KMeans(n_clusters=k, n_init='auto').fit(cancer.data).inertia_ for k in range(2, 7)]
plt.plot(range(2, 7), inertia, marker='o')
plt.title("34. Elbow Plot (Inertia)")
plt.show()

# 35. Concentric circles, Agglomerative Clustering with single linkage.
X, _ = make_circles(n_samples=300, factor=0.5, noise=0.05)
agg = AgglomerativeClustering(n_clusters=2, linkage='single').fit(X)
plt.scatter(X[:, 0], X[:, 1], c=agg.labels_)
plt.title("35. Single Linkage on Circles")
plt.show()

# 36. Wine dataset, DBSCAN after scaling, count clusters (excluding noise).
X_scaled = StandardScaler().fit_transform(wine.data)
dbscan = DBSCAN(eps=2.5, min_samples=5).fit(X_scaled)
n_clusters = len(set(dbscan.labels_)) - (1 if -1 in dbscan.labels_ else 0)
print(f"36. Number of clusters: {n_clusters}")

# 37. make_blobs and KMeans. Plot cluster centers on top of data.
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)
kmeans = KMeans(n_clusters=3, n_init='auto').fit(X)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            c='red', marker='X', s=200, label='Centers')
plt.title("37. KMeans with Centroids")
plt.show()

# 38. Load Iris, cluster with DBSCAN, print noise count.
dbscan = DBSCAN(eps=0.5, min_samples=5).fit(iris.data)
print(f"38. Noise count: {list(dbscan.labels_).count(-1)}")

# 39. make_moons, apply K-Means, visualize results.
X, _ = make_moons(n_samples=200, noise=0.05, random_state=42)
kmeans = KMeans(n_clusters=2, n_init='auto').fit(X)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.title("39. K-Means on Moons")
plt.show()

# 40. Digits dataset, PCA to 3 components, KMeans, 3D scatter plot.
X_pca3 = PCA(n_components=3).fit_transform(digits.data)
kmeans = KMeans(n_clusters=10, n_init='auto').fit(X_pca3)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca3[:, 0], X_pca3[:, 1], X_pca3[:, 2], c=kmeans.labels_, cmap='tab10')
plt.title("40. 3D KMeans Digits")
plt.show()

# 41. Blobs with 5 centers, KMeans, evaluate with silhouette_score.
X, _ = make_blobs(n_samples=500, centers=5, random_state=42)
kmeans = KMeans(n_clusters=5, n_init='auto').fit(X)
print(f"41. Silhouette Score: {silhouette_score(X, kmeans.labels_):.3f}")

# 42. Breast Cancer, PCA reduction, Agglomerative Clustering (2D).
X_pca2 = PCA(n_components=2).fit_transform(cancer.data)
agg = AgglomerativeClustering(n_clusters=2).fit(X_pca2)
plt.scatter(X_pca2[:, 0], X_pca2[:, 1], c=agg.labels_)
plt.title("42. Agglomerative (PCA 2D)")
plt.show()

# 43. Noisy circular data, KMeans vs DBSCAN side-by-side.
X, _ = make_circles(n_samples=500, factor=0.5, noise=0.1)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.scatter(X[:, 0], X[:, 1], c=KMeans(n_clusters=2, n_init='auto').fit_predict(X))
ax1.set_title("KMeans")
ax2.scatter(X[:, 0], X[:, 1], c=DBSCAN(eps=0.15).fit_predict(X))
ax2.set_title("DBSCAN")
plt.show()

# 44. Iris dataset, plot Silhouette Coefficient for each sample (KMeans).
kmeans = KMeans(n_clusters=3, n_init='auto').fit(iris.data)
vals = silhouette_samples(iris.data, kmeans.labels_)
y_lower = 10
for i in range(3):
    ith_vals = np.sort(vals[kmeans.labels_ == i])
    size = ith_vals.shape[0]
    y_upper = y_lower + size
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_vals)
    y_lower = y_upper + 10
plt.title("44. Silhouette Coefficients (Iris)")
plt.show()

# 45. make_blobs, Agglomerative Clustering ('average' linkage).
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
agg = AgglomerativeClustering(n_clusters=4, linkage='average').fit(X)
plt.scatter(X[:, 0], X[:, 1], c=agg.labels_)
plt.title("45. Average Linkage Blobs")
plt.show()

# 46. Wine dataset, KMeans, seaborn pairplot (first 4 features).
df = pd.DataFrame(wine.data[:, :4], columns=wine.feature_names[:4])
df['cluster'] = KMeans(n_clusters=3, n_init='auto').fit_predict(wine.data)
sns.pairplot(df, hue='cluster')
plt.suptitle("46. Wine Cluster Pairplot", y=1.02)
plt.show()

# 47. Noisy blobs, DBSCAN cluster and noise count.
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)
dbscan = DBSCAN(eps=1.0, min_samples=5).fit(X)
labels = dbscan.labels_
print(f"47. Clusters: {len(set(labels)) - (1 if -1 in labels else 0)}, Noise: {list(labels).count(-1)}")

# 48. Digits, t-SNE reduction, Agglomerative Clustering.
X_tsne = TSNE(n_components=2, random_state=42).fit_transform(digits.data)
agg = AgglomerativeClustering(n_clusters=10).fit(X_tsne)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=agg.labels_, cmap='tab10')
plt.title("48. Agglomerative (t-SNE)")
plt.show()
```
